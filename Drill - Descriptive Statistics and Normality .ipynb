# Drill - Descriptive Statistics and Normality

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
normal = np.random.normal(1, 3, 100)
chisquare = np.random.chisquare(1, 100)
gumbel = np.random.gumbel(1, 3, 100)
exp = np.random.exponential(2, 100)
gamma = np.random.gamma(1, 3, 100)
pareto = np.random.pareto(1, 100)
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.hist(normal, bins=20, color='c')
plt.axvline(normal.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normal.mean() + normal.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(normal.mean() - normal.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Normal')

plt.subplot(1,3,2)
plt.hist(chisquare, bins=20, color='c')
plt.axvline(chisquare.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(chisquare.mean() + chisquare.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(chisquare.mean() - chisquare.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Chi Square')

plt.subplot(1,3,3)
plt.hist(gumbel, bins=20, color='c')
plt.axvline(gumbel.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(gumbel.mean() + gumbel.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(gumbel.mean() - gumbel.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Gumbel')

plt.show()
plt.tight_layout()


plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.hist(exp, bins=20, color='c')
plt.axvline(exp.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exp.mean() + exp.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exp.mean() - exp.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Exponential')

plt.subplot(1,3,2)
plt.hist(gamma, bins=20, color='c')
plt.axvline(gamma.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(gamma.mean() - gamma.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Gamma')

plt.subplot(1,3,3)
plt.hist(pareto, bins=20, color='c')
plt.axvline(pareto.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(pareto.mean() + pareto.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(pareto.mean() - pareto.std(), color='b', linestyle='dashed', linewidth=2)
plt.title('Pareto') 

plt.show()
plt.tight_layout()
```

Mean/Standard Deviation is least useful in Chi Square and Pareto. However, they are useful in helping to describe the Normal, Gumbel, Exponential and Gamma distributions. Using mean and standard deviation, you will be able to get an indicator of how many values (or percentage of values) fall within a certain range. 

```python
rand1 = np.random.normal(5,0.5,100)
rand2 = np.random.normal(10,1,100)
rand3 = rand1 + rand2 

plt.hist(rand3, bins=20, color='c')
plt.axvline(rand3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean() + rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean() - rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()

print(rand3.mean())
rand3.std()
```

The mean of the third variable is similar to the first two variables added together. The standard deviation of the third variable does not seem to correlate with the first two variables.
